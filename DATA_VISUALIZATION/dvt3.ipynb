# Task
Generate Python code to create stacked bar charts, grouped bar charts, segmented bar charts, and mosaic plots for visualizing the relationship between categorical variables.

## Prepare the data

### Subtask:
Select the categorical columns for visualization.


**Reasoning**:
Identify and list the categorical columns from the DataFrame for subsequent visualization.



```python
categorical_cols = ['Location', 'Has_Garage', 'Has_Garden']
print(categorical_cols)
```

## Create stacked bar charts

### Subtask:
Generate stacked bar charts to show the distribution of one categorical variable within each category of another.


**Reasoning**:
Generate stacked bar charts to visualize the distribution of `Has_Garage` and `Has_Garden` within each `Location`.



```python
fig, axes = plt.subplots(1, 1, figsize=(12, 5))

# Stacked bar chart for Has_Garden vs Location
garden_location = df.groupby(['Location', 'Has_Garden']).size().unstack()
garden_location.plot(kind='bar', stacked=True, ax=axes)
axes.set_title('Distribution of Has_Garden by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Count')
axes.legend(title='Has_Garden', labels=['No Garden', 'Has Garden'])
axes.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

## Create grouped bar charts

### Subtask:
Generate grouped bar charts to compare the distribution of one categorical variable across the categories of another.


**Reasoning**:
Generate the grouped bar charts as requested in the instructions.



```python
fig, axes = plt.subplots(1, 1, figsize=(6, 5))

garden_location = df.groupby(['Location', 'Has_Garden']).size().unstack()

garden_location.plot(kind='bar', ax=axes, color=["#1f77b4", "#2ca02c"])

axes.set_title('Distribution of Has_Garden by Location')   # Updated title
axes.set_xlabel('Location')
axes.set_ylabel('Count')
axes.legend(title='Has_Garden', labels=['No Garden', 'Has Garden'])
axes.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

## Create segmented bar charts

### Subtask:
Generate segmented bar charts to visualize the proportion of one categorical variable within each category of another.


**Reasoning**:
Generate segmented bar charts to visualize the proportion of one categorical variable within each category of another.



```python
fig, axes = plt.subplots(1, 1, figsize=(6, 5))

# Segmented bar chart for Has_Garden vs Location
garden_location = df.groupby(['Location', 'Has_Garden']).size().unstack()
garden_location_prop = garden_location.divide(garden_location.sum(axis=1), axis=0)
garden_location_prop.plot(kind='bar', stacked=True, ax=axes)
axes.set_title('Proportion of Has_Garden by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Proportion')
axes.legend(title='Has_Garden', labels=['No Garden', 'Has Garden'])
axes.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

## Create mosaic plots

### Subtask:
Generate mosaic plots to display the relationship between two or more categorical variables.


**Reasoning**:
Import the mosaic function and create mosaic plots for the specified categorical variable relationships with titles.



```python
from statsmodels.graphics.mosaicplot import mosaic

# Create mosaic plot for Location and Has_Garden
plt.figure(figsize=(8, 6))
mosaic(df, ['Location', 'Has_Garden'], title='Mosaic Plot of Location and Has_Garden')
plt.show()
```

*   Grouped bar charts were created to compare the absolute counts of houses with and without a garage, and with and without a garden, across different locations.
*   Segmented bar charts were generated to visualize the proportion of houses with and without a garage, and with and without a garden, within each `Location`. These charts provide a relative comparison across locations.
*   Mosaic plots were successfully generated using `statsmodels` to display the relationship between `Location` and `Has\_Garage`, and between `Location` and `Has\_Garden`, providing a visual representation of the conditional proportions.

### Insights or Next Steps

# Task
Analyze the relationship between two continuous variables using a scatter plot with a fit line and the relationship between a categorical and a continuous variable using a bar chart of means, a grouped kernel density plot, a box plot, a violin plot, and a beeswarm plot.

## Continuous vs continuous analysis

### Subtask:
Create a scatter plot with a regression line using two continuous variables.


**Reasoning**:
Generate a scatter plot with a regression line using two continuous variables from the dataframe `df`.



```python
sns.regplot(x='Size_sqm', y='Price', data=df)
plt.title('Scatter plot of Size_sqm vs Price with Regression Line')
plt.xlabel('Size (sqm)')
plt.ylabel('Price')
plt.show()
```

## Categorical vs continuous analysis

### Subtask:
Create a bar chart showing the mean of a continuous variable grouped by a categorical variable.


**Reasoning**:
Generate a bar chart showing the mean of the continuous variable 'Price' grouped by the categorical variable 'Location'.



```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Location', y='Price', data=df)
plt.title('Mean Price by Location')
plt.xlabel('Location')
plt.ylabel('Mean Price')
plt.show()
```

## Categorical vs continuous analysis

### Subtask:
Generate a grouped kernel density plot for a continuous variable across categories of a categorical variable.


**Reasoning**:
Generate a grouped kernel density plot for a continuous variable across categories of a categorical variable as requested in the instructions.



```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Size_sqm', hue='Location', fill=True)
plt.title('Kernel Density Estimate of Size_sqm by Location')
plt.xlabel('Size (sqm)')
plt.ylabel('Density')
plt.show()
```

## Categorical vs continuous analysis

### Subtask:
Create a box plot for a continuous variable across categories of a categorical variable.


**Reasoning**:
Generate a box plot to visualize the distribution of the continuous variable 'Price' across the categories of the categorical variable 'Location'.



```python
fig, axes = plt.subplots(1, 1, figsize=(8, 6))
sns.boxplot(x='Location', y='Price', data=df, ax=axes)
axes.set_title('Box plot of Price by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Price')
plt.show()
```

## Categorical vs continuous analysis

### Subtask:
Generate a violin plot for a continuous variable across categories of a categorical variable.


**Reasoning**:
Generate a violin plot to visualize the distribution of the continuous variable 'Price' across the categories of the categorical variable 'Location'.



```python
fig, axes = plt.subplots(1, 1, figsize=(8, 6))
sns.violinplot(x='Location', y='Price', data=df, ax=axes)
axes.set_title('Violin plot of Price by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Price')
plt.show()
```

## Categorical vs continuous analysis

### Subtask:
Create a beeswarm plot for a continuous variable across categories of a categorical variable.


**Reasoning**:
Generate a beeswarm plot using seaborn to visualize the distribution of 'Price' for each 'Location'.



```python
fig, axes = plt.subplots(1, 1, figsize=(10, 6))
sns.swarmplot(x='Location', y='Price', data=df, ax=axes)
axes.set_title('Beeswarm plot of Price by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Price')
plt.show()
```

*   A bar chart showed the mean `Price` for each `Location`.
*   A grouped kernel density plot visualized the distribution of `Size_sqm` across different `Location` categories.
*   A box plot displayed the distribution of `Price` across different `Location` categories, highlighting quartiles and outliers.
*   A violin plot showed the distribution shape of `Price` for each `Location`.
*   A beeswarm plot displayed the individual data points for `Price` within each `Location` category.
