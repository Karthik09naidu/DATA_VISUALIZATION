```python
import pandas as pd
import plotly.express as px

# Load the synthetic dataset CSV
df = pd.read_csv('/content/t5sales_data (1).csv')

# TreeMap
fig_treemap = px.treemap(
    df,
    path=['Root', 'Region', 'Category', 'Subcategory'],
    values='Sales',
    color='Sales',
    color_continuous_scale='RdBu',
    title='Sales TreeMap by Region, Category, and Subcategory'
)
fig_treemap.show()

# Sunburst
fig_sunburst = px.sunburst(
    df,
    path=['Root', 'Region', 'Category', 'Subcategory'],
    values='Sales',
    color='Sales',
    color_continuous_scale='RdBu',
    title='Sales Sunburst by Region, Category, and Subcategory'
)
fig_sunburst.show()
```
